CBOW处理训练集和测试集
输入：处理好的pkl文件
输出：词向量文件
处理方法：BOW

```
from sklearn.feature_extraction.text import CountVectorizer
import pickle
import numpy  as np
from sklearn.decomposition import PCA
from sklearn.random_projection import GaussianRandomProjection
from sklearn.random_projection import SparseRandomProjection
from sklearn import manifold
```

```
#导入训练集文件
with open("/content/drive/My Drive/Colab Notebooks/数据集/security_train.pkl", "rb") as f:   # 加载PKL文件，测试数据
   train_files = pickle.load(f)
print("Train_PKL_File Loading Finish")
with open("/content/drive/My Drive/Colab Notebooks/数据集/security_test.pkl", "rb") as f:   # 加载PKL文件，测试数据
    test_files = pickle.load(f)
print("Test_PKL_File Loading Finish")
```

BOW：CountVectorizer参数说明

ngram_range：tuple (min_n, max_n)
要提取的不同n-gram的n值范围的下边界和上边界。 将使用n的所有值，使得min_n <= n <= max_n

stop_words = None  不使用英语停用词

max_df： float in range [0.0, 1.0] or int, default=1.0

在构建词汇表时，忽略文档频率严格高于给定阈值的术语（语料库特定的停用词）。 如果是float，则参数表示文档的比例，整数绝对计数。 如果词汇表不是None，则忽略此参数。


min_df：float in range [0.0, 1.0] or int, default=1

构建词汇表时，请忽略文档频率严格低于给定阈值的术语。 该值在文献中也称为截止值。 如果是float，则参数表示文档的比例，整数绝对计数。 如果词汇表不是None，则忽略此参数。


```
#BOW处理训练集
print("start BOW...")
vectorizer = CountVectorizer(ngram_range=(1,4),stop_words = None, min_df=0.0, max_df=1.0,)  # tf-idf特征抽取ngram_range=(1,5)
train_features = vectorizer.fit_transform(train_files)          # 将api长序列进行TFIDF，作为样本特征  每行api组合为一个长文本序列  抽取长文本的特征 生成特征向量 
#以上每行”label+长文本（api组合）“转换为了“label+特征向量（每行特征向量数目不同，原因是因为所包含API名称的数目不同 所以抽取出来的特征数也不一样）”
print("BOW Finish!")
```

```
train_features
```

```
#降维
print("demension reduction start!")
#高斯降维
transformer = GaussianRandomProjection(n_components=500)
train_features = transformer.fit_transform(train_features)
#PCA
#transformer = SparseRandomProjection(n_components=500, random_state=0)
#train_features = transformer.fit_transform(train_features )
#transformer=PCA(n_components=500)
#train_features=transformer.fit(train_features)
print("demension reduction end!")   
print(train_features.shape) 
```

```
#保存处理后的文件
with open("/content/drive/My Drive/Colab Notebooks/code/BOW_XGB_1102/bow_feature_train1102.pkl", 'wb') as f:  # pickle
    pickle.dump(train_features, f)     #序列化对象  将对象保存到文件中
with open("/content/drive/My Drive/Colab Notebooks/code/BOW_XGB_1102/bow_feature_train1102.pkl", 'rb') as f:
    train_features = pickle.load(f)    #反序列化  
print(train_features.shape)
print(" PKL Finish!")
```
